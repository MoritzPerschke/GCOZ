```python
import json
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

# Basic stuff
import data

```python
infile = open("../ROTTR_trial/data_1.json");
data = json.load(infile)
infile.close()

baseline = pd.DataFrame(data["baseline"]).add_suffix("%")
all_slowed = pd.DataFrame(data["all"]).add_suffix("%")
present = pd.DataFrame(data["method8"]).add_suffix("%")
```

# Look at baseline Data

```python
baseline.describe().apply(lambda s: s.apply('{0:.1f}'.format)).transpose()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(15, 15));

# raw data
sb.scatterplot(baseline, ax=axs[0,0])
baseline.plot(ax=axs[0,1], kind="box")

# exclude 95th percentile
sb.scatterplot(baseline[baseline < baseline.quantile(.95)], ax=axs[1,0])
baseline[baseline < baseline.quantile(.95)].plot(ax=axs[1,1], kind="box", showfliers=False)
```

# All methods slowed down
